<h1>Step 1 : Create source embeddings for the text column

```python
from sentence_transformers import SentenceTransformer
encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df.Material_Description)
```

```python
vectors.shape
```

```python
dim = vectors.shape[1]
dim
```

<h2>Step 2 : Build a FAISS Index for vectors

```python
import faiss

index = faiss.IndexFlatL2(dim)
```

<h2>Step 3 : Normalize the source vectors (as we are using L2 distance to measure similarity) and add to the index

```python

index.add(vectors)
index
```

<h1>Step 4 : Encode search text using same encorder and normalize the output vector

```python
search_query = "show the products with CAPACITY 800 KG"
vec = encoder.encode(search_query)
import numpy as np
svec = np.array(vec).reshape(1,-1)
svec.shape
faiss.normalize_L2(svec)
svec.shape
```

<h1>Step 5: Search for similar vector in the FAISS index created

```python
distances, I = index.search(svec, k=2)
distances
```
